####  1. Identify which items are produced the most.  
First let's take a look at which items are produced the most across all countries and all years. Let's reexamine our dataframe.

```python
filleddf.head()
```

Finding the top thirty most produced food items

```python
filleddf.groupby("Item")["Element"].agg("count").sort_values(ascending=False)[:30]
```

The top three items produced around the world are "Eggs", "Cereals-Excluding Beer", and "Fish, Seafood".  

```python
%matplotlib inline
items=pd.DataFrame(filleddf.groupby("Item")["Element"].agg("count").sort_values(ascending=False))[:30]
items.plot(kind='barh', figsize=(10,15), width=0.7, color=[('#ff9999','#66b3ff','#99ff99','#ffcc99', '#cc99ff', '#99ffff', '#ffff99', '#00c060', '#e7fff3', '#c0d8cc')], legend=False)
plt.gca().invert_yaxis()
plt.title("Top 30 Items Produced of Food/Feed Around the World", fontsize="15")
plt.xlabel("Number of Food Item", fontsize="12")
plt.ylabel("Item (Food/Feed)", fontsize="12")
plt.grid()
plt.show()
```

#### 2. Identify the countries that produce the most food.
Next let's look at the countries that produce the most food across all years and all food items.  

Saving all years into col_years variable:

```python
col_years=[col for col in filleddf if col.startswith('Y')]
```

Grouping dataframe by area and doing a sum for each year.

```python
top_countries=filleddf.groupby("Area")[col_years].sum()
```

```python
top_countries
```

Getting average of tons for all years.  

```python
top_allyears=pd.DataFrame(top_countries.agg("mean",axis=1).sort_values(ascending=False),columns=["Tons"])
```

Looking at top five values.  

```python
top_allyears.head()
```

```python
top_allyears[:30].plot(kind="barh", width=0.7, color=[('#ff9999','#66b3ff','#99ff99','#ffcc99', '#cc99ff', '#99ffff', '#ffff99', '#00c060', '#e7fff3', '#c0d8cc')], figsize=(10,15), legend=False)
plt.gca().invert_yaxis()
plt.title("Average Tons Produced of Food/Feed by Country/Territory", fontsize="15")
plt.xlabel("1000s of Tons", fontsize="12")
plt.ylabel("Country/Territory", fontsize="12")
plt.grid()
plt.show()
```

"China, mainland" produced the most amount of food and feed in 1000s of tons followed by the United States and India.  South Africa produces the least amount of food.  

#### 3. The food production trends over time
We can now look at the top 10 countries that produce the most food over all years.

```python
top_countries_sorted=top_countries.sort_values(by=['Y2013'], ascending=False)
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#cc99ff', '#99ffff', '#ffff99', '#00c060', '#e7fff3', '#c0d8cc']
top_countries_sorted[:10].T.plot(color=colors, figsize=(15,9), linewidth=2.5)
plt.title("Top Ten Countries that Produce the Most Food/Feed from 1961 to 2013", fontsize="15")
plt.xlabel("Year", fontsize="12")
plt.ylabel("Food Produced", fontsize="12")
plt.grid()
plt.show()
```

It is interesting to note from this graph that India has surpassed the U.S. in production in the mid-2000s.  China has the most production up to 2013 which is where this data stops.
